```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from functools import partial
from typing import Callable, Optional
```

```python
def central_limit_theorem(sample_function: Callable[[int], np.ndarray], 
                          samples: int, 
                          sample_size: int, 
                          title: Optional[str] = None, 
                          plot: bool = True) -> np.ndarray:
    
    means = [sample_function(sample_size).mean() for _ in range(samples)]
    
    if plot:
        if title is not None:
            plt.title(title)
        sns.histplot(means, kde=True )

    return means
```

```python
sample_size = 10000
trials = 1000
```

```python
# Uniform distribution

low = 0
high = 100

uniform_fn = partial(np.random.uniform, low, high)
uniform_means = central_limit_theorem(uniform_fn, trials, sample_size, "Uniform Distribution")
```

```python
# Normal distribution

mean = 0
std = 1

normal_fn = partial(np.random.normal, mean, std)

normal_means = central_limit_theorem(normal_fn, trials, sample_size, "Normal Distribution")
```

```python
# Exponential distribution

scale = 1

exponential_fn = partial(np.random.exponential, scale)

exponential_means = central_limit_theorem(exponential_fn, trials, sample_size, "Exponential Distribution")
```
