```python
import torch
from torch.distributions import Categorical, Exponential
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def clt(samplef,trials,sample_size,name, cat=None):
    means = []

    if cat is not None:
        for i in range(trials):
            means.append(torch.mean(numbers[samplef(torch.Size([sample_size]))]).item())
    else:
        for i in range(trials):
            means.append(torch.mean(samplef(torch.Size([sample_size]))).item())
    
    plt.title(name)
    sns.histplot(means, kde=True );
```

```python
sample_size = 10000
trials = 1000
```

```python
numbers = torch.arange(0,100,1/100)
categorical = Categorical(numbers)
clt(categorical.sample, trials, sample_size,"categorical", cat = numbers)
```
